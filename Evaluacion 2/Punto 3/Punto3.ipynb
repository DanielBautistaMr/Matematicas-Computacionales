```python
import numpy as np
import matplotlib.pyplot as plt

x = [1, 3, 5, 7, 9]
y = [2.8, 3, 3.5, 3.8, 4.5]

def lagrange_interpolation(x, y, degree, x_interp):
    n = len(x)
    if degree > n - 1:
        raise ValueError("El grado del polinomio de interpolaci贸n no puede ser mayor que el n煤mero de puntos - 1.")
    
    result = 0
    for i in range(degree + 1):
        term = y[i]
        for j in range(degree + 1):
            if i != j:
                term *= (x_interp - x[j]) / (x[i] - x[j])
        result += term
    
    return result

x_interp = 4.75

grados = [1, 2, 3, 4]

estimaciones = []
for grado in grados:
    estimacion = lagrange_interpolation(x, y, grado, x_interp)
    estimaciones.append(estimacion)

for grado, estimacion in zip(grados, estimaciones):
    print(f"Estimaci贸n de f(4.75) del polinomio de grado {grado}: {estimacion}")

plt.figure(figsize=(12, 10))

# Grado 1
plt.subplot(2, 2, 1)
plt.scatter(x, y, color='red', label='Datos')
plt.plot(x,y, color='red', alpha=0.2)
plt.plot(x_interp, estimaciones[0], 'bo', label='Grado 1')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polinomio de grado 1')
plt.legend()

# Grado 2
plt.subplot(2, 2, 2)
plt.scatter(x, y, color='red', label='Datos')
plt.plot(x,y, color='red', alpha=0.2)
plt.plot(x_interp, estimaciones[1], 'go', label='Grado 2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polinomio de grado 2')
plt.legend()

# Grado 3
plt.subplot(2, 2, 3)
plt.scatter(x, y, color='red', label='Datos')
plt.plot(x,y, color='red', alpha=0.2)
plt.plot(x_interp, estimaciones[2], 'mo', label='Grado 3')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polinomio de grado 3')
plt.legend()

# Grado 4
plt.subplot(2, 2, 4)
plt.scatter(x, y, color='red', label='Datos')
plt.plot(x,y, color='red', alpha=0.2)
plt.plot(x_interp, estimaciones[3], 'co', label='Grado 4')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polinomio de grado 4')
plt.legend()

plt.tight_layout()
plt.show()
```
