```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de los puntos
x = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y = np.array([3.8, 4, 4.2, 4.6, 4.9, 5.3, 5.7, 6.1])

# Ajuste exponencial
exp_fit = np.polyfit(x, np.log(y), 1)
exp_func = np.exp(exp_fit[1]) * np.exp(exp_fit[0] * x)

# Ajuste de potencias
power_fit = np.polyfit(np.log(x), np.log(y), 1)
power_func = np.exp(power_fit[1]) * x**power_fit[0]

# Ajuste de la raz√≥n de crecimiento
growth_rate = np.diff(y) / np.diff(x)
avg_growth_rate = np.mean(growth_rate)
growth_func = y[0] * (avg_growth_rate**(x - 1))

# Gr√°fica del ajuste exponencial
print("La mejor aproximaci√≥n visualmente parece ser el AJUSTE EXPONENCIAL.")
plt.figure(figsize=(8, 6))
plt.scatter(x, y, label='Puntos')
plt.plot(x, exp_func, label='Ajuste Exponencial',color='red', alpha=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.title('Ajuste Exponencial')
plt.show()

# Gr√°fica del ajuste de potencias
plt.figure(figsize=(8, 6))
plt.scatter(x, y, label='Puntos')
plt.plot(x, power_func, label='Ajuste de Potencias',color='red', alpha=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.title('Ajuste de Potencias')
plt.show()

# Gr√°fica del ajuste de la raz√≥n de crecimiento
plt.figure(figsize=(8, 6))
plt.scatter(x, y, label='Puntos')
plt.plot(x, growth_func, label='Ajuste Raz√≥n de Crecimiento',color='red', alpha=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.title('Ajuste Raz√≥n de Crecimiento')
plt.show()


```
